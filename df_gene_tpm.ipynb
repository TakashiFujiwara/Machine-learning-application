```python
df=df.dropna()
```

```python
df_raw_data=df.iloc[:,3:111]
```

```python
df_gene_length=df["gene_length"]
```

```python
df_gene_length.head()
```

```python
def totalreads_normalize(df):
    return 10**6*df/df.sum()

def genelength_normalize(df, gene_length):
    return (df.T*10**3/gene_length).T
```

```python
df_temp=genelength_normalize(df_raw_data, df_gene_length)
df_tpm=totalreads_normalize(df_temp)
df_tpm.to_csv('tpm.csv', sep=',', header=True)
```
